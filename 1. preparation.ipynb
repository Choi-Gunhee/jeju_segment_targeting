# load library

```python
# 파일 경로 패턴 정의 및 읽기를 위한 라이브러리
import pandas as pd
import glob
import re

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# load data

##### 여행 선호도 기대값 데이터 불러오기

```python
# 파일 경로 패턴 정의
pattern = "data/domestic_travel_area_inferrence/CI_DMSTC_TOUR_AREA_INTRST_DGREE_INFO_*.csv"

# 모든 CSV 파일 읽기
all_files = glob.glob(pattern)

# 데이터프레임 리스트
dataframes = []

# 각 파일을 읽고 데이터프레임 리스트에 추가
for file in all_files:
    df = pd.read_csv(file)
    dataframes.append(df)

# 모든 데이터프레임을 하나로 합치기
combined_df1 = pd.concat(dataframes, ignore_index=True)

# 결과 출력
#print(combined_df.head())
combined_df1
```

##### 여행지 및 소비액 데이터 불러오기

```python
# 파일 경로 패턴 정의
pattern = "data/domestic_travel_cost/CI_TOUR_CT_EXPNDTR_CHARTR_INFO_*.csv"


# 모든 CSV 파일 읽기
all_files = glob.glob(pattern)

# 데이터프레임 리스트
dataframes = []

# 각 파일을 읽고 데이터프레임 리스트에 추가
for file in all_files:
    df = pd.read_csv(file)
    dataframes.append(df)

# 모든 데이터프레임을 하나로 합치기
combined_df2 = pd.concat(dataframes, ignore_index=True)

# 결과 출력
#print(combined_df2.head())
combined_df2
```

# data preprocessing

##### 여행 선호 데이터프레임과 소비 데이터프레임 병합

```python
combined_df = pd.merge(combined_df1, combined_df2, on=['RESPOND_ID','EXAMIN_BEGIN_DE','SEXDSTN_FLAG_CD','AGRDE_FLAG_NM','HSHLD_INCOME_DGREE_NM'], how='left')
combined_df
```

##### 칼럼 명 변경

```python
column_names = {'SEXDSTN_FLAG_CD':'SEX','AGRDE_FLAG_NM':'AGE','ANSWRR_OC_AREA_NM':'AREA','HSHLD_INCOME_DGREE_NM':'INCOME',
                'SU_TOUR_INTRST_VALUE':'SEOUL','GG_IC_TOUR_INTRST_VALUE':'GYEONGGI','BS_TOUR_INTRST_VALUE':'BUSAN','CHN_DJ_TOUR_INTRST_VALUE':'CHUNGNAM',
                'CB_TOUR_INTRST_VALUE':'CHUNGBUK','GN_US_TOUR_INTRST_VALUE':'GYEONGNAM', 'GB_TOUR_INTRST_VALUE':'GYEONGBUK','JN_GJ_TOUR_INTRST_VALUE':'JEONNAM',
                'JNBK_TOUR_INTRST_VALUE':'JEONBUK','GW_TOUR_INTRST_VALUE':'GANGWAN','JJ_TOUR_INTRST_VALUE':'JEJU','GB_DG_TOUR_INTRST_VALUE':'GYEONGBUK',
                'TOUR_TOT_CT_VALUE':'Total','TOUR_LDGMNT_CT_VALUE':'Lodgement','TOUR_FOOD_CT_VALUE':'Food','TOUR_TRNSPORT_CT_VALUE':'Transport',
                'TOUR_SHOPNG_CT_VALUE':'Shopping','TOUR_ACTVTY_CT_VALUE':'Activity','TOUR_ETC_CT_VALUE':'Etc',
                'TOUR_CTPRVN_NM':'City','TOUR_SIGNGU_NM':'District','TOUR_COM_NMPR_NM':'Company','TOUR_PD_VALUE':'Per_Diem','MRRG_AT_NM':'Married',
                'CHLDRN_TY_NM':'Children','OCCP_NM':'Occupation'}

combined_df.rename(columns=column_names, inplace=True)
```

##### 관심도 수치화

```python
# 데이터프레임의 '관심도' 컬럼의 값 수치화
response = ['많이 적어졌다', '약간 적어졌다', '예전과 비슷하다', '약간 커졌다' , '많이 커졌다']

for index, res in enumerate(response):
    combined_df.replace(res, index+1, inplace=True)
```

##### 소비액 수치화

##### '모름' 데이터 NA로 변경

```python
combined_df['Total'].replace('모름', np.nan, inplace=True)
```
